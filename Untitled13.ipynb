```python
import numpy as np
import matplotlib.pyplot as plt

def initial_parameters():
    # initialize parameters within specified ranges with random values
    dA = np.random.randint(5, 21)
    dB = np.random.randint(5, 21)
    rA = round(np.random.uniform(0.1, 0.9), 2)
    rB = round(np.random.uniform(0.1, 0.9), 2)
    number_of_units = 100
    number_of_cycles = 10000

    # create initial inventory
    inventory = [
        [number_of_units // 2, dA, rA],
        [number_of_units // 2, dB, rB],
        number_of_units
    ]

    unsatisfactory_demand = [0, 0]  # initialize unsatisfied demand counters
    redistribution = [3, 0]  # initialize redistribution values

    return [inventory, unsatisfactory_demand, redistribution, 0, number_of_cycles]

def simulate_cycle(cycle_parameters):
    inventory, unsatisfactory_demand, redistribution, cycle_it, number_of_cycles = cycle_parameters
    inven_A, dA, rA = inventory[0]
    inven_B, dB, rB = inventory[1]
    rng = np.random.default_rng()

    # generate daily random demand
    cycle_dA = np.random.uniform(0, dA)
    cycle_dB = np.random.uniform(0, dB)

    if cycle_dA > inven_A:
        unsatisfactory_demand[0] += 1
    if cycle_dB > inven_B:
        unsatisfactory_demand[1] += 1

    # update inventory
    new_A = rng.binomial(inven_A, rA)
    new_B = rng.binomial(inven_B, rB)

    if new_A <= redistribution[0] or new_B <= redistribution[0]:
        new_A, new_B = inventory[2] // 2, inventory[2] // 2
        redistribution[1] += 1

    # update inventory
    inventory[0][0] = new_A
    inventory[1][0] = new_B
    cycle_it += 1

    # return updated parameters
    return [inventory, unsatisfactory_demand, redistribution, cycle_it, number_of_cycles]

def monte_carlo_simulation(cycle_para, redistribution_val=None):
    if redistribution_val is None:
        redistribution_val = cycle_para[2][0]

    # run simulation
    number_of_cycles = cycle_para[4]
    for _ in range(number_of_cycles):
        cycle_para = simulate_cycle(cycle_para)

    P_unsatisfactoryA = cycle_para[1][0] / number_of_cycles
    P_unsatisfactoryB = cycle_para[1][1] / number_of_cycles
    redistribution_rate = cycle_para[2][1] / number_of_cycles

    # return results
    return cycle_para, P_unsatisfactoryA, P_unsatisfactoryB, redistribution_rate

def redistribution_changes(cycle_para, lower_val, upper_val):
    results = []

    # run simulations for each redistribution value
    for redistribution_val in range(lower_val, upper_val + 1):
        simulation_result = monte_carlo_simulation(cycle_para, redistribution_val)
        results.append((redistribution_val, *simulation_result[1:]))

    return results

def main_simulation():
    cycle_para = initial_parameters()

    # run Monte Carlo simulation
    cycle_para, P_unsatisfactoryA, P_unsatisfactoryB, redistribution_rate = monte_carlo_simulation(cycle_para)
    print(f"Unsatisfied demand rate for A: {P_unsatisfactoryA}")
    print(f"Unsatisfied demand rate for B: {P_unsatisfactoryB}")
    print(f"Redistribution rate: {redistribution_rate}")

    results = redistribution_changes(cycle_para, 1, 20)

    # extract data
    ks = [result[0] for result in results]
    P_unsatisfactoryAs = [result[1] for result in results]
    P_unsatisfactoryBs = [result[2] for result in results]
    redistribution_rates = [result[3] for result in results]

    # plot results
    plt.figure(figsize=(12, 6))
    plt.plot(ks, redistribution_rates, marker='o', label='Redistribution Rate')
    plt.plot(ks, P_unsatisfactoryAs, marker='x', label='Unsatisfied Demand Rate for A')
    plt.plot(ks, P_unsatisfactoryBs, marker='^', label='Unsatisfied Demand Rate for B')
    plt.xlabel('Redistribution Value (k)')
    plt.ylabel('Rates')
    plt.title('Rates vs Redistribution Value')
    plt.legend()
    plt.grid(True)
    plt.show()

main_simulation()
```
